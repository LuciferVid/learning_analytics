# Exploratory Data Analysis: Students Performance in Exams

This notebook explores the dataset to understand the features and prepare for model development.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/StudentsPerformance.csv')
df.head()
```

## Visualizations

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
plt.show()
```

```python
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['percentage'] = df['total_score'] / 3

def get_grade(score):
    if score >= 80: return 'High-performing'
    if score >= 50: return 'Average'
    return 'At-risk'

df['performance_category'] = df['percentage'].apply(get_grade)
df['performance_category'].value_counts()
```
